```python
# webdrvier-manager 라이브러리 불러오기
from webdriver_manager.chrome import ChromeDriverManager
ChromeDriverManager().install()

from selenium import webdriver
browser = webdriver.Chrome()

from selenium.webdriver.common.by import By
```

```python
url = "https://www.yes24.com/Product/Category/BestSeller?categoryNumber=001&pageNumber=1&pageSize=24"
browser.get(url)
# get_attribute 메서드는 Selenium WebDriver에서 웹 요소의 특정 속성 값을 가져오는 데 사용됩니다. 웹 요소의 속성은 HTML 태그에 정의된 속성들로, id, class, name, href, src 등이 포함
browser.find_element(By.CLASS_NAME, 'gd_nameE').get_attribute('href') 

datas = browser.find_elements(By.CLASS_NAME, 'gd_nameE')
for i in datas:
    print(i.get_attribute('href'))
```

```python
# 3페이지까지의 링크 데이터 수집
from webdriver_manager.chrome import ChromeDriverManager
ChromeDriverManager().install()

from selenium import webdriver
browser = webdriver.Chrome()

from selenium.webdriver.common.by import By
import time

# 페이지를 리스트에 담기
link_list = []
for i in range(1,4):
    url = f'https://www.yes24.com/Product/Category/BestSeller?categoryNumber=001&pageNumber={i}&pageSize=24'
    browser.get(url)
    browser.find_element(By.CLASS_NAME, 'gd_name').get_attribute('href') 
    
    datas = browser.find_elements(By.CLASS_NAME, 'gd_name')

    for i in datas:
        link = i.get_attribute('href')
        link_list.append(link)

    time.sleep(3)
print(link_list)
```

```python
# 데이터베이스 연동 후 -> 수집한 데이터를 DB에 저장 (csv)
import pymysql

conn = pymysql.connect(
    host = 'localhost',
    user = 'root',
    password = 'Clzlsfjqm8*',
    db = 'yes24',
    charset = 'utf8mb4',
    cursorclass = pymysql.cursors.DictCursor
)

# execute를 실행하기위함
with conn.cursor() as cur:
    
    for link in link_list:
        browser.get(link)
        
        title = browser.find_element(By. CLASS_NAME,'gd_name').text
        author = browser.find_element(By. CLASS_NAME,'gd_auth').text
        publisher = browser.find_element(By. CLASS_NAME,'gd_pub').text
        publishing = browser.find_element(By. CLASS_NAME,'gd_date').text
        rating = browser.find_element(By.CLASS_NAME,'yes_b').text
        reviews = browser.find_element(By.CLASS_NAME,'txC_blue').text
        sales = browser.find_element(By.CLASS_NAME,'gd_sellNum').text.split(" ")[2]
        price = browser.find_element(By.CLASS_NAME,'yes_m').text[:-1]
        ranking = browser.find_element(By.CLASS_NAME,'gd_best').text.split(" | ")[0]
        # ranking_weeks = browser.find_element(By.CLASS_NAME,'gd_best').text.split(" | ")[1]

        sql = """INSERT INTO books(
            title, author, publishing, rating, reviews, sales, price, ranking, ranking_weeks
            )
            VALUES(
            %s, %s, %s, %s, %s, %s, %s, %s, %s
            )
            """
        

        cur.execute(sql, (title, author, publishing, rating, reviews, sales, price, ranking))
        conn.commit()
```
